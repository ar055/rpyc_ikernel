```python
import time 
last_time = time.time()

from maix import display, camera
from PIL import Image, ImageDraw

display.local_show(False)
display.hook(clear_output=True)

while True:
    im = camera.capture()
    if im:
        draw = ImageDraw.Draw(im)
        last_time = time.time()
        draw.text((10, 10), str((1000 // (1000 * (time.time() - last_time)))), (255, 0, 0))
        display.show(im)
```
