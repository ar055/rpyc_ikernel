```python
from maix import display, camera
from PIL import Image, ImageDraw
import time
display.local_show(False)
display.hook(clear_output=True)
last_time = time.time()
while True:
#     print(time.asctime())
    try:
        im = camera.capture()
        if im:
            draw = ImageDraw.Draw(im)
            fps = 1000 // (1000 * (time.time() - last_time))
            last_time = time.time()
            draw.text((10, 10), str(fps), (255, 0, 0))
            display.show(im)
    except Exception as e:
        print(e)
```
